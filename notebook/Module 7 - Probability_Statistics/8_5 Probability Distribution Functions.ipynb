## Probability Distribution (and density) functions
   
A probability density function is a function that may be used to define a particular **probability distribution**. 

Additionally, the probability distriubtion function is a function whose value at any given sample, **x**, in the sample space can be interpreted as providing a relative likelihood that the value of the random variable would equal that sample.

<img src="images/probability_distribution.jpg" width="300">

The two classes of such a distribution are **discrete and continuous**. The former represented by a **probability mass function** (discrete) and the latter by a **probability density function** (continuous).



## Discrete Probability Distribution

A probability distribution is a function that tells us how probable different outcomes are in an experiment. It describes events in terms of their probabilities; this is out of all possible outcomes. 

Examples of probability distributions for **discrete** variables (probability density) include:
 - Binomial
 - Poisson
 - Bernoulli

## Continuous Probability Distribution

A continuous distribution describes the probabilities of the possible values of a continuous random variable. A continuous random variable is a random variable with a set of possible values (known as the range) that is infinite and uncountable.

Examples of probability distributions for continiuous variables (probability density):
 - Normal
 - Chi-squared 

```python
# Normal
import matplotlib.pyplot as plt
import numpy as np

# normal distribution with a mean centered at 90, 
# with a standard deviation of 2, and 10000 (ten thousand) 
# random data points created.
mu, sigma = 90, 2
s = np.random.normal(mu, sigma, 10000)

# Create the bins and histogram
count, bins, ignored = plt.hist(s, 20, normed=True)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=3, color='y')
plt.xlabel('Normal')
plt.ylabel('Frequency')
plt.show()
```

Chi-Square test is a statistical method to determine if two categorical variables have a significant correlation between them. Both those variables should be from the same population and they should be categorical like  Yes/No, Male/Female, Red/Green, etc.

```python
# chi squared distribution
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
fig,ax = plt.subplots(1,1)

linestyles = [':', '--', '-.', '-']
deg_of_freedom = [1, 4, 7, 6]
for df, ls in zip(deg_of_freedom, linestyles):
  ax.plot(x, stats.chi2.pdf(x, df), linestyle=ls)

plt.xlim(0, 10)
plt.ylim(0, 0.4)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Chi-Square Distribution')

plt.show()
```
