## What is the difference between Probability and Statistics?
   
Probability and statistics are related areas of mathematics which are focused on analyzing the relative frequency of events. However, there is a key difference related to the approach:
- **Probability**: the study of probability deals with predicting the liklihood of **future** events
- **Statistics**: the study of statistics deals with analyzing the frequency of **past** events

Statistical theory measures the extent to which the world is ideal (what has happened).

Probability theory enables the discovery of the consequences of the ideal world (what "may" happen)

## Statistics in Python - Coin toss example

**Probability** seeks to answer the question, “What is the chance of an event happening?” An event is some outcome of interest. 

To calculate the chance of an event happening, we also need to consider all the other events that can occur.

The quintessential example of probability is the **coin toss**. In a coin toss the only events that can happen are:

- **Flipping a heads**

- **Flipping a tails**

These two events are the set of all possible events and represent the **sample space**. 

To calculate the probability of an event occurring, we sum how many times an event of interest can occur and divide it by the **sample space** value. 

For example, a fair coin will have a 1-in-2 chance of being heads or tails.

However, if we were to perform the experiment of flipping a coin, we’re likely to get an irregular count of heads or tails every once in a while. 

We can use statistics to calculate probabilities based on experimental observations from the real world and check how it compares to the ideal (the probability formula).

Let's try an example of using experimentation and statistics to gather data to support the conjecture that flipping a coin will yield a 50-50 probability of either outcome.

```python
import random
def coin_trial():
    heads = 0
    for i in range(100):
        if random.random() <= 0.5:
            heads +=1
    return heads

def simulate(n):
   trials = []
   for i in range(n):
       trials.append(coin_trial())
   return(sum(trials)/n)
```

```python
# 10, 100, 1000, 10000, 100000 coin flips - what is the average %
print("10: ", simulate(10))
print("100: ", simulate(100))
print("1000: ", simulate(1000))
print("10000: ", simulate(10000))
print("100000: ", simulate(100000))
```

## Probability in Python - Coin toss example

The **binomial distribution** model deals with finding the probability of success of an event which has only **two possible outcomes** in a series of experiments. 

For example, tossing of a coin always gives a head or a tail (two outcomes). 

We can use the binomial distribution to determine the probability of finding exactly **``n``** heads in tossing a coin repeatedly for **count** times.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function that runs our coin toss trials
# heads is a list of the number of successes from each trial of # tosses
def run_simulated_coin_toss(trials, num_tosses):
    heads = []
    for i in range(trials):
        tosses = [np.random.random() for i in range(num_tosses)]
        heads.append(len([i for i in tosses if i>=0.50]))
    return heads
```

```python
# there are a few deprecated features
# we will ignore the warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Run the function
heads = run_simulated_coin_toss(10000, 10) 
# Plot the results as a histogram
fig, ax = plt.subplots(figsize=(14,7))
ax = sns.distplot(heads, bins=11, label='simulation results', kde=False)
ax.set_xlabel("Number of Heads",fontsize=16)
ax.set_ylabel("Frequency (# trials)",fontsize=16)
plt.legend()
plt.show()
```

```python
# Plot the actual binomial distribution as a sanity check
from scipy.stats import binom
x = range(0,11)# range is 0 to 10
n = 10  # the number of trials
p = 0.5 # the probability of a heads

# Plot the results as a histogram
fig, ax1 = plt.subplots(figsize=(14,7))
# ax1 = sns.distplot(heads, bins=11, label='simulation results')
ax1.set_xlabel("Number of Heads",fontsize=16)
ax1.set_ylabel("Probabilty",fontsize=16)
# draw the actual binomal distribution
ax1.plot(x, binom.pmf(x, n, p), 'r-', label='actual binomial distribution')
ax1.vlines(x, 0, binom.pmf(x, n, p), colors='r', lw=5, alpha=0.5)
plt.legend()
plt.show()
```

```python
# Probability of getting 6 heads
runs = 10000
prob_6 = sum([1 for i in np.random.binomial(n, p, size=runs) 
              if i==6])/runs
print('The probability of 6 heads is: ' + str(prob_6))
```
